## 1.1 Random number generators

Random number generators (RNGs) are the backbone of computational statistics. There are two fundamental types: true random number generators based on physical processes, and pseudo-random number generators (PRNGs), which are deterministic algorithms designed to mimic randomness.

True randomness comes from nature. Atomic decay is perhaps the most quoted example. You can check random.org for many examples. These are slow and fundamentally unpredictable. In contrast, computational PRNGs like linear congruential generators (LCGs) are fast, reproducible, and deterministic; you feed in a seed and always get the same sequence.

Replicability, in particular, is a crucial aspect for benchmarking many computations. However, poorly designed PRNGs can have hidden structure. Let us investigate this here. Consider the RANDU generator with the following recursion relation
$$
I_{i+1} = (65539\, I_i)\; \mathrm{mod}\; 2^{31},
$$
which needs to start from an odd integer. One can then generate PRNGs floating point numbers in the interval $[0,1]$ via the following mapping:
$$
u_i = \frac{I_i}{2^{31}}
$$

a. Implement the RANDU generator using the recurrence.

b. Generate $1000$ successive pairs $(u_{2i}, u_{2i+1})$ and plot them in the unit square. Comment on whether the points appear uniformly distributed.

c. Choose a small window in the unit square and keep sampling until you collect roughly $1000$ points in that window. Plot and discuss any visible structure.

d. Repeat the same zoom test with numpy's random number generator, compare the two plots, and briefly explain what this shows about RNG quality.
