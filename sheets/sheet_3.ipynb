# 3.1 Photon counts (analytical)

In X-ray astronomy, a detector counts
individual photons. The number of counts $N_i$ recorded during the $i$-th
observation interval of duration $t_i$ follows a Poisson distribution:

$$P(N_i = k \mid \lambda) = \frac{(\lambda t_i)^k \, e^{-\lambda t_i}}{k!},$$

where $\lambda > 0$ is the unknown source count rate (photons per second).
You have $n$ independent observations $(N_1, t_1), \ldots, (N_n, t_n)$.

a. Write down the log-likelihood ${L}(\lambda)$, dropping all terms that do not depend on $\lambda$.

b. Derive the MLE $\hat{\lambda}_\mathrm{MLE}$ analytically and give a physical interpretation.

c. Show that $\hat{\lambda}_\mathrm{MLE}$ is unbiased using $\mathbb{E}[N_i] = \lambda t_i$.

d. Compute the Fisher information $F(\lambda)$ using either of the forms in Eq. (4.19) of the notes. Verify they agree.
*Hint:* for a Poisson variable $\mathrm{Var}(N_i) = \lambda t_i$.

# 3.2 Fitting an emission line and pitfalls with numerical derivatives

Emission and absorption lines are arguably the most important observables in
astrophysics. Consider a spectrum of $M$ wavelength bins with flux
measurements $\{d_k\}$ at wavelengths $\{\lambda_k\}$:

$$d_k = A\exp\!\left(-\frac{(\lambda_k - \mu)^2}{2\sigma_\ell^2}\right)
+ B + \varepsilon_k, \qquad \varepsilon_k \sim \mathcal{N}(0, \sigma_n^2),$$

where the noise level $\sigma_n = 0.5$ is **known**, and
$\theta = (A,\, \mu,\, \sigma_\ell,\, B)^\top$ are amplitude, line centre,
line width, and continuum level. Note that the only source of randomness is the noise term $\varepsilon_k$. The true parameters are: $\theta_0  = (5.0, 6563.0, 3.0, 1.0)^\top$.


a. Write down the log-likelihood ${L}(\theta)$ and show that maximising it is
equivalent to least squares here.

b. Read in the data points from `./data/spectral_data.txt` and implement the MLE by minimising $-{L}(\theta)$ with
`scipy.optimize.minimize`. Use multiple starting points and verify you
always recover the same solution.

c. Implement the score $s_\alpha(\theta) = \partial_\alpha {L}(\theta)$ first analytically. Then implement forward and central differences:

$$s_\alpha^\mathrm{fwd} \approx
\frac{{L}(\theta_\alpha + h_\alpha) - {L}(\theta)}{h_\alpha},
\qquad
s_\alpha^\mathrm{cen} \approx
\frac{{L}(\theta_\alpha + h_\alpha)
     - {L}(\theta_\alpha - h_\alpha)}{2h_\alpha},$$
here $h_\alpha$ is a perturbation to the $\alpha$-th parameter.

d. For each parameter, plot $|s_\alpha^\mathrm{num} - s_\alpha^\mathrm{exact}|/s_\alpha^\mathrm{exact}$
against $h \in [10^{-12}, 10^{0}]$. Explain the shape of the curve: what
sets the optimal $h$ and why does performance degrade on each side?

e. Repeat for the diagonal elements of the Hessian
$\partial_\alpha\partial_\alpha \mathcal{L}$ (which enter the Fisher matrix via Eq. 4.19).
How does the optimal $h$ shift compared to the gradient case?

f. Estimate the parameter covariance matrix from the Hessian of
$-{L}$ at $\hat{\theta}_\mathrm{MLE}$ numerically and give the uncertainty for each parameter $\sigma_{\hat{\theta}_\mathrm{MLE}}$. 

g. Set $\sigma_n = 5$ so the continuum $B$ and amplitude $A$ become poorly determined (low-SNR regime). Compute the full Fisher matrix via finite differences and compare element-by-element to the analytical result, what do you observe?
