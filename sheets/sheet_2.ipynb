## 2.1 Bias and variance in the exponential distribution

Let $X_1, \ldots, X_n$ be i.i.d. random variables drawn from an exponential distribution with rate parameter $\lambda > 0$:

$$f(x \mid \lambda) = \lambda e^{-\lambda x}, \quad x \geq 0.$$

a. Show that the true mean of this distribution is $\mu = 1/\lambda$ and the true variance is $\sigma^2 = 1/\lambda^2$.

b. Show that the sample mean $\hat{\mu} = \frac{1}{n}\sum_{i=1}^n X_i$ is an unbiased estimator of $\mu = 1/\lambda$, and compute its variance $\sigma^2_{\hat{\mu}}$.

c. Somebody proposes the following estimator for $\lambda$:

$$\hat{\lambda} = \frac{1}{\hat{\mu}} = \frac{n}{\sum_{i=1}^n X_i}.$$

Argue whether $\hat{\lambda}$ is biased or unbiased. In which direction does the bias go?

d. Compute the Mean Squared Error (MSE) of $\hat{\mu}$ as an estimator of $\mu$. For an unbiased estimator, the MSE reduces to the variance — verify that this is the case here and interpret the result physically.


# 2.2 The central limit theorem

In the lecture (Section 3.6), the CLT was stated for arbitrary distributions with finite variance. In this exercise you will verify it numerically for the Poisson distribution.

A Poisson random variable $X \sim \text{Pois}(\mu)$ ($\sim$ reads as "follows") has the probability mass function

$$P(X = k) = \frac{\mu^k e^{-\mu}}{k!}, \qquad k = 0, 1, 2, \ldots,$$

with mean $\mu$ and variance $\mu$.

Write a Python script that does the following:

a. Fix $\mu = 3$. For each of $n \in \{1, 2, 5, 20, 100\}$, draw $N_{\rm samples} = 10{,}000$ realisations of the standardised sum

$$S_n = \frac{1}{\sqrt{n}} \sum_{i=1}^{n} \frac{X_i - \mu}{\sqrt{\mu}},$$

where each $X_i \sim \text{Pois}(\mu)$ independently. You can use *numpy* to sample from a Poisson distribution.

b. For each value of $n$, plot a normalised histogram of $S_n$ and overlay the standard normal $\mathcal{N}(0, 1)$ PDF.

c. Quantify the convergence by computing, for each $n$, the **sample kurtosis** of $S_n$:

$$\kappa_4 = \frac{E[(S_n - \bar{S}_n)^4]}{\left(E[(S_n - \bar{S}_n)^2]\right)^2}$$

and plot $\kappa_4$ vs $n$ on a log-linear scale. Caclualte the kurtosis for a normal distribution. Does your result converge to this value? What rate of convergence do you observe?
